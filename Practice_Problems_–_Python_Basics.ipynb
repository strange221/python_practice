Write a Python program that takes an integer as input and calculates the sum of its digits

```python
def sum_of_digits(number):
    number_str = str(abs(number))
    digit_sum = 0
    for digit_char in number_str:
        digit_sum += int(digit_char)

    return digit_sum

try:
    user_input = input("Enter an integer: ")
    input_number = int(user_input)
    result = sum_of_digits(input_number)

    print(f"\nThe integer you entered is: {input_number}")
    print(f"The sum of its digits is: {result}")

except ValueError:
    print("\nInvalid input. Please enter a whole number (integer).")
```

Create a Python program to calculate the factorial of a given number using a function.

```python
def calculate_factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1:
        return 1

    else:
        result = 1

        for i in range(2, n + 1):
            result *= i

        return result

try:
    user_input = input("Enter a non-negative integer to calculate its factorial: ")
    number = int(user_input)
    factorial_result = calculate_factorial(number)
    print(f"\nThe number entered is: {number}")

    if isinstance(factorial_result, int):
        print(f"The factorial of {number} is: {factorial_result}")
    else:
        print(f"Error: {factorial_result}")

except ValueError:
    print("\nInvalid input. Please enter a whole number (integer).")
```

Develop a Python program that checks whether a given word or phrase is a palindrome.

```python
import re

def is_palindrome(text):


    text = text.lower()

    cleaned_text = re.sub(r'[^a-z0-9]', '', text)

    return cleaned_text == cleaned_text[::-1]

user_input = input("Enter a word or phrase to check for palindrome: ")

is_it_a_palindrome = is_palindrome(user_input)

print(f"Input: '{user_input}'")

if is_it_a_palindrome:
    print("Result: It IS a palindrome!")
else:
    print("Result: It is NOT a palindrome.")
```

Create a Python program that takes a sentence as input and counts the number of vowels and consonants.

```python
def count_vowels_and_consonants(sentence):
    vowels = {'a', 'e', 'i', 'o', 'u'}

    vowel_count = 0
    consonant_count = 0
    sentence = sentence.lower()

    for char in sentence:
        if 'a' <= char <= 'z':
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    return vowel_count, consonant_count

user_input = input("Enter a sentence or a word: ")

v, c = count_vowels_and_consonants(user_input)

print(f"Input Text: '{user_input}'")
print(f"Total Vowels: {v}")
print(f"Total Consonants: {c}")
print(f"Total Letters Counted: {v + c}")
```

Develop a Python program that takes two lists as input and returns a new list containing the common elements between them.

```python
def find_common_elements(list1, list2):
    set1 = set(list1)
    common_elements = []
    for element in list2:
        if element in set1:
            common_elements.append(element)
            set1.remove(element)

    return common_elements
list_a = [1, 2, 3, 4, 5, 6, 6]
list_b = [5, 6, 7, 8, 9, 1]

result_list = find_common_elements(list_a, list_b)

print(f"List 1: {list_a}")
print(f"List 2: {list_b}")
print(f"Common Elements: {result_list}")
```

Write a Python program that takes a string as input and counts the frequency of each character in the string.

```python
def count_character_frequency(input_string):
    frequency = {}
    for char in input_string:
        frequency[char] = frequency.get(char, 0) + 1

    return frequency
user_input = input("Enter a string (word or sentence): ")

char_counts = count_character_frequency(user_input)
print(f"Original String: '{user_input}'")
print("\nCharacter Counts:")
for char, count in char_counts.items():
    print(f"  {repr(char):<5}: {count}")
```

Develop a Python program that checks if a given year is a leap year.

```python
def is_leap_year(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

try:
    user_input = input("Enter a year: ")
    input_year = int(user_input)
    if input_year < 1:
        print("Please enter a positive year (AD/CE).")
    else:
        if is_leap_year(input_year):
            print(f"\nResult: {input_year} IS a leap year!")
        else:
            print(f"\nResult: {input_year} is NOT a leap year.")

except ValueError:
    print("\nInvalid input. Please enter a whole number for the year.")
```

Write a Python program that checks if a given number, is a prime number.

```python
import math

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    limit = int(math.sqrt(n))
    for i in range(3, limit + 1, 2):
        if n % i == 0:
            return False
    return True

try:
    user_input = input("Enter a positive integer to check for primality: ")
    number = int(user_input)

    if is_prime(number):
        print(f"\nResult: {number} IS a prime number!")

    else:
        print(f"\nResult: {number} is NOT a prime number.")

except ValueError:
    print("\nInvalid input. Please enter a positive whole number (integer).")
```

Write a Python program to generate the Fibonacci sequence up to a specified number of terms.

```python
def generate_fibonacci(n_terms):
    if n_terms <= 0:
        return "Please enter a positive integer for the number of terms."

    if n_terms == 1:
        return [0]
    sequence = [0, 1]
    for _ in range(2, n_terms):
        next_term = sequence[-1] + sequence[-2]
        sequence.append(next_term)

    return sequence

try:
    user_input = input("Enter the number of Fibonacci terms you want to generate: ")
    num_terms = int(user_input)

    fib_sequence = generate_fibonacci(num_terms)
    print(f"Number of terms requested: {num_terms}")

    if isinstance(fib_sequence, list):
        print("The sequence is:")
        print(f"{' -> '.join(map(str, fib_sequence))}")
    else:
        print(f"Error: {fib_sequence}")

except ValueError:
    print("\nInvalid input. Please enter a whole number (integer).")
```

Write a python program to find the roots of a quadratic equation

```python
import math
def solve_quadratic_equation(a, b, c):
    if a == 0:
        if b == 0:
            return "Invalid equation: Both 'a' and 'b' are zero. The equation is c = 0."
        root = -c / b
        return f"This is a linear equation. The single root is x = {root:.4f}"
    discriminant = b**2 - 4 * a * c
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2 * a)
        root2 = (-b - math.sqrt(discriminant)) / (2 * a)

        return (f"The equation has two distinct real roots:\n"
                f"  Root 1 (x1): {root1:.4f}\n"
                f"  Root 2 (x2): {root2:.4f}")
    elif discriminant == 0:
        root = -b / (2 * a)
        return (f"The equation has one real, repeated root:\n"
                f"  Root (x): {root:.4f}")
    else:
        real_part = -b / (2 * a)
        imaginary_part = math.sqrt(abs(discriminant)) / (2 * a)

        return (f"The equation has two distinct complex roots:\n"
                f"  Root 1 (x1): {real_part:.4f} + {imaginary_part:.4f}i\n"
                f"  Root 2 (x2): {real_part:.4f} - {imaginary_part:.4f}i")

try:
    a = float(input("Enter coefficient 'a' (for x^2): "))
    b = float(input("Enter coefficient 'b' (for x): "))
    c = float(input("Enter coefficient 'c' (constant): "))
    result = solve_quadratic_equation(a, b, c)
    print("\n" + "="*30)
    print(f"Equation: {a}x^2 + {b}x + {c} = 0")
    print("="*30)
    print(result)

except ValueError:
    print("\nInvalid input. Please ensure all coefficients are valid numbers.")
```
