Write a Python program to accept a string value from the user and display the count of each character in that string. Sample input: Enter a string value: assembly Sample output: a=1, s=2, e=1, m=1, b=1, l=1, y=1

```python
def count_and_display_frequency(input_string):
    frequency = {}
    for char in input_string:
        frequency[char] = frequency.get(char, 0) + 1
    output_parts = []
    for char, count in frequency.items():
        output_parts.append(f"{char}={count}")
    return ", ".join(output_parts)

user_input = input("Enter a string value: ")

frequency_output = count_and_display_frequency(user_input)

print("Sample output:", frequency_output)
```

Write a Python function to find the maximum of three numbers Sample input: 34,12,7 Sample output: 34

```python
def find_maximum(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

try:
    user_input = input("Enter three numbers separated by commas (e.g., 34,12,7): ")

    numbers_str = [s.strip() for s in user_input.split(',')]

    if len(numbers_str) != 3:
        print("Error: You must enter exactly three numbers.")
    else:
        n1 = float(numbers_str[0])
        n2 = float(numbers_str[1])
        n3 = float(numbers_str[2])

        max_value = find_maximum(n1, n2, n3)
        print(f"\nSample input: {n1},{n2},{n3}")
        print(f"Sample output: {max_value}")

except ValueError:
    print("\nInvalid input. Please ensure all values entered are valid numbers.")
```

Write a Python function called exponent(base,exp) that returns an integer value of base raises to the power of exp. Sample input: Enter the base: 2 Enter the exponent: 3 Sample output: 8

```python
def exponent(base, exp):
    result = base ** exp
    return result

try:
    base_input = input("Enter the base: ")
    base_value = float(base_input)

    exp_input = input("Enter the exponent: ")
    exp_value = int(exp_input)

    result = exponent(base_value, exp_value)

    print(f"\nSample input: Enter the base: {base_input}, Enter the exponent: {exp_input}")
    if result == int(result):
        display_result = int(result)
    else:
        display_result = result

    print(f"Sample output: {display_result}")

except ValueError:
    print("\nInvalid input. Please ensure you enter valid numbers for the base and exponent.")
```

Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number. Sample input: 4 Sample output: 36.

```python
def sum_of_cubes_smaller_than(n):
    if not isinstance(n, int) or n <= 1:
        return 0

    total_sum = 0

    for i in range(1, n):
        total_sum += i ** 3

    return total_sum

try:
    user_input = input("Enter a positive integer: ")
    number = int(user_input)

    if number <= 1:
        result = 0
    else:
        result = sum_of_cubes_smaller_than(number)
    print(f"\nSample input: {number}")
    print(f"Calculation: 1^3 + ... + ({number}-1)^3")
    print(f"Sample output: {result}")

except ValueError:
    print("\nInvalid input. Please enter a whole positive number (integer).")
```

Write a Python program which iterates from 1 to 10. For multiples of 2, print “Fizz” instead of the number and for the multiples of 5, print “Buzz”. For numbers which are multiples of both 2 and 5, print “FizzBuzz”. Sample input: numbers from 1 to 10 Sample output: 1 Fizz 3 Fizz Buzz Fizz 7 Fizz 9 FizzBuzz

```python
def run_fizzbuzz(start, end):
    output_list = []
    for number in range(start, end + 1):

        output = ""

        if number % 2 == 0:
            output += "Fizz"

        if number % 5 == 0:
            output += "Buzz"
        if not output:
            output = str(number)

        output_list.append(output)

    return " ".join(output_list)


START_NUM = 1
END_NUM = 10

fizzbuzz_sequence = run_fizzbuzz(START_NUM, END_NUM)

print(f"Sample input: numbers from {START_NUM} to {END_NUM}")
print("Sample output:", fizzbuzz_sequence)
print("\nFormatted Output:")
for item in fizzbuzz_sequence.split():
    print(item)
```

Write a Python program to find the most frequent item in a list of numbers. Sample input: 2, 3, 4, 2, 5, 2 Sample output: 2

```python
from collections import Counter

def find_most_frequent(data_list):

    if not data_list:
        return None
    counts = Counter(data_list)

    most_common_item_tuple = counts.most_common(1)

    return most_common_item_tuple[0][0]


try:
    user_input = input("Enter a list of numbers separated by commas (e.g., 2, 3, 4, 2, 5, 2): ")

    input_list = [float(s.strip()) for s in user_input.split(',')]
    cleaned_list = [int(x) if x == int(x) else x for x in input_list]

    most_frequent_value = find_most_frequent(cleaned_list)

    print(f"\nSample input: {', '.join(map(str, cleaned_list))}")
    print(f"Sample output: {most_frequent_value}")


except ValueError:
    print("\nInvalid input. Please ensure all values entered are valid numbers.")
```

Write a Python program to find the sum of squares of the numbers in a list. Sample input: 2,1,3,1 Sample output: 15

```python
def sum_of_squares(number_list):

    total_sum = 0

    for number in number_list:

        square = number ** 2
        total_sum += square

    return total_sum

try:
    user_input = input("Enter a list of numbers separated by commas (e.g., 2,1,3,1): ")

    input_list = [float(s.strip()) for s in user_input.split(',')]
    cleaned_list = [int(x) if x == int(x) else x for x in input_list]

    result = sum_of_squares(cleaned_list)

    print(f"\nSample input: {', '.join(map(str, cleaned_list))}")
    print(f"Calculation: 2² + 1² + 3² + 1² = 4 + 1 + 9 + 1 = 15")
    print(f"Sample output: {result}")

except ValueError:
    print("\nInvalid input. Please ensure all values entered are valid numbers.")
```

Write a Python program using for loop that will iterate from 1 to 15. For each iteration, check if the current number is odd or even, and display the message to the screen as odd or even. Sample input: 1….15 Sample output: 1-odd 2-even …. 15-odd

```python
def check_odd_or_even(start, end):
    print(f"Sample input: {start}....{end}")
    print("\nSample output:")
    for number in range(start, end + 1):
        if number % 2 == 0:
            result = "even"
        else:
            result = "odd"

        print(f"{number}-{result}")

START = 1
END = 15

check_odd_or_even(START, END)
```

Write a Python program to convert temperatures to and from Celsius Fahrenheit. [Formula: c/5=f-32/9 where c=temperature in Celsius and f= temperature in Fahrenheit.] Sample input: Temperature in Fahrenheit =41 Sample output: Temperature in Celsius =5

```python
def celsius_to_fahrenheit(celsius):

    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit):

    return (fahrenheit - 32) * 5/9

try:
    value = float(input("Enter the temperature value: "))

    unit = input("Is this temperature in Celsius (C) or Fahrenheit (F)? ").strip().upper()

    print("\n" + "="*40)

    if unit == 'F':

        celsius_result = fahrenheit_to_celsius(value)

        print(f"Sample input: Temperature in Fahrenheit ={value}")
        print(f"Sample output: Temperature in Celsius ={round(celsius_result)}")

    elif unit == 'C':
        # Convert Celsius to Fahrenheit
        fahrenheit_result = celsius_to_fahrenheit(value)

        # Display the result
        print(f"Sample input: Temperature in Celsius ={value}")
        print(f"Sample output: Temperature in Fahrenheit ={round(fahrenheit_result)}")

    else:
        print("Invalid unit entered. Please enter 'C' or 'F'.")

    print("="*40)

except ValueError:
    print("\nInvalid temperature input. Please enter a valid number.")
```

Write a Python function to calculate the factorial of a number (a nonnegative integer). The function accepts the number as an argument. Sample input: 3 Sample output: 6

```python
def calculate_factorial(n):

    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")

    if n == 0 or n == 1:
        return 1

    result = 1

    for i in range(2, n + 1):
        result *= i

    return result


try:

    user_input = input("Enter a non-negative integer: ")
    number = int(user_input)

    factorial_result = calculate_factorial(number)

    print(f"\nSample input: {number}")
    print(f"Sample output: {factorial_result}")

except ValueError as e:
    print(f"\nError: {e}. Please enter a whole non-negative number.")
```
