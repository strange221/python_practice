14) Group by Director: avg Rating, Revenue

```python
director_stats = md.groupby('Director').agg(
    Avg_Rating=('Rating', 'mean'),
    Avg_Revenue=('Revenue (Millions)', 'mean')
)
director_stats_sorted = director_stats.sort_values(by='Avg_Rating', ascending=False)
director_stats_sorted
```

15) Group by Year: count, avg Rating

```python
year_stats = md.groupby('Year').agg(
    Movie_Count=('Title', 'count'),
    Avg_Rating=('Rating', 'mean')
)
year_stats
```

16) Create a new column Profitability:

a) Revenue divided by Duration

b) Find top 10 most profitable movies

```python
md['Profitability'] = md['Revenue (Millions)'] / md['Runtime (Minutes)']
top_10_profitable = md.nlargest(n=10, columns='Profitability')
top_10_profitable
```

17) Create a column that categorizes movies by rating:

a) Rating < 6: Poor

b) 6  Rating < 7.5: Average

c) Rating  7.5: Excellent

```python
bins = [0, 6, 7.5, 10.1]
labels = ['Poor', 'Average', 'Excellent']
md['Rating_Category'] = pd.cut(
    x=md['Rating'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)
md
```

18) Create a custom function to score movies based on:

i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
def calculate_high_score(row):
  rating = row['Rating']
  revenue = row['Revenue (Millions)']
  high_score = (rating * 10) + revenue
  return high_score
md['High score'] = md.apply(calculate_high_score, axis=1)
md
```
