# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_range(data_list):
  minimum_value = data_list[0]
  #setting the first value of the list as minimum and maximum value
  maximum_value = data_list[0]

  for num in data_list:
# looping through the entire list to check for
# maximum_value and minimum_value
    if (num >= maximum_value):
      maximum_value = num
    elif (num <= minimum_value):
      minimum_value = num
  return(minimum_value, maximum_value)


mylist = [5,6,7,56,20,33,11]
range = find_range(mylist)
range
```

## Q2. Write function that returns mean value of a list of numbers

```python
def calculate_mean(data_list):
  total = 0
  for i in data_list:
    total += i
  return total / len(data_list)

mylist = {2,3,4,5,7,6,10}
print(f"mean = {calculate_mean(mylist)}")
```
