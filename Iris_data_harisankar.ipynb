```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
plt.plot(iris['sepal_length'])
plt.plot(iris['sepal_width'])
plt.plot(iris['petal_length'])
plt.plot(iris['petal_width'])

plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Line Plot of Iris Features")
plt.show()
```

```python
plt.plot(iris['petal_length'],color="black",marker='o',linestyle='dashed',markerfacecolor='pink',markersize=10)
```

```python
plt.plot(iris['petal_width'],color='green')

avg_sepal_length = np.mean(iris['sepal_width'])

plt.axhline(y=avg_sepal_length, color='black', linestyle='--')

plt.xlabel('x values')
plt.ylabel('Sepal width')
plt.title('Sepal width with Average Line')
plt.legend(['Sepal width', 'Average Sepal width'])
plt.show()
```

```python
plt.bar(iris["species"], iris["sepal_length"],color="blue")
plt.xlabel("Species")
plt.ylabel(" Sepal Length")
plt.title(" Sepal Length of Iris Species")
plt.show()
```

```python
sepal_width = iris.groupby("species")["sepal_width"].first()
petal_width  = iris.groupby("species")["petal_width"].first()
plt.bar(sepal_width.index,sepal_width.values, color="orange")
plt.bar(petal_width.index, petal_width.values, color="blue")
```

```python
iris_species=iris['species'].value_counts()
labels=iris_species.index
plt.pie(iris_species,labels=labels)
plt.show()
```
